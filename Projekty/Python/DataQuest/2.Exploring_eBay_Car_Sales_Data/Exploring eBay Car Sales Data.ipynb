## 2. Cleaning Column Names

1. Use the DataFrame.columns attribute to print an array of the existing column names.

2. Copy that array and make the following edits to columns names:
    - **yearOfRegistration** to **registration_year**
    - **monthOfRegistration** to **registration_month**
    - **notRepairedDamage** to **unrepaired_damage**
    - **dateCreated** to **ad_created**
    - The rest of the column names from camelcase to snakecase.

```python
copy_columns = autos.columns
```

```python
mapping_columns = {
    'yearOfRegistration' : 'registration_year',
    'monthOfRegistration' : 'registration_month',
    'notRepairedDamage' : 'unrepaired_damage',
    'dateCreated' : 'ad_created',
}
```

*Generate large letters*

```python
uni_large_chr = list(range(65,91))
large_letters = []
for u in uni_large_chr:
    large_letters.append(chr(u))
```

```python
print(large_letters)
```

*Functions to check if the column has large letters and also transform name of column to the **snakecase***

```python
def check_letters(large_letters,column_name):
    for i in large_letters:
        if i in column_name:
            return True

def pos_large_letter(large_letters,column_name):
    positions = []
    for i in large_letters:
        if i in column_name:
            index = column_name.find(i)
            positions.append(index)
    if len(positions) == 1:
        result = (column_name[:positions[0]],'_',column_name[positions[0]:])
        result = ''.join(result).lower()
    else:
        result = (column_name[:positions[0]],'_',column_name[positions[0]:positions[1]],'_',column_name[positions[1]:])
        result = ''.join(result).lower()
    return result
```

```python
proper_columns = {}
for column in copy_columns:
    if column in list(mapping_columns.keys()):
        proper_columns[column] = mapping_columns[column]
    elif check_letters(large_letters,column):
        proper_columns[column] = pos_large_letter(large_letters,column)
    else:
        proper_columns[column] = column

print(proper_columns)
```

3. Assign the modified column names back to the DataFrame.columns attribute.

```python
autos.columns = autos.columns.map(proper_columns)
```
