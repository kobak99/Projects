## 2. Data Cleaning

1. Isolate the `Time` and the `US_dollar` columns. Assign them to a different variable named `euro_to_dollar`.

```python
exchange_rates = exchange_rates[['Period\\Unit:','[US dollar ]']]
```

```python
# We rename the [US dollar ] and Period\Unit: columns to something easier to type — US_dollar and Time.
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
```

```python
# We change the Time column to a datetime data type.
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
# Convert the US_dollar column to a float data type.
exchange_rates['US_dollar'] = exchange_rates['US_dollar'].replace('-', np.nan)
exchange_rates['US_dollar'] = exchange_rates['US_dollar'].astype(float)
```

```python
# We sort the values by Time in ascending order.
exchange_rates.sort_values('Time', inplace=True)
# We reset the index (and drop the initial index).
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates
```

2. Replace `null` values with the average of two periods, the previous one and the next one.

```python
# How much nulls we have
euro_to_dollar.isnull().sum()
```

```python
# Interpolate values
euro_to_dollar['US_dollar'].interpolate(inplace=True)
```

## 3. Rolling Mean

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

1. Calculate the rolling means for the `US_dollar` column using a moving window of 30 days. Add the rolling means to a new column named `rolling_mean`.

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
plt.plot(euro_to_dollar.loc[(euro_to_dollar['Time'] >= '2010-01-01'),'Time'],
         euro_to_dollar.loc[(euro_to_dollar['Time'] >= '2010-01-01'),'US_dollar'])
plt.show()
```

```python
plt.plot(euro_to_dollar.loc[(euro_to_dollar['Time'] >= '2010-01-01'),'Time'],
         euro_to_dollar.loc[(euro_to_dollar['Time'] >= '2010-01-01'),'rolling_mean'])
plt.show()
```

## 4. Coming up with an Idea

We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), Donald Trump (2017-2021) and Joe Biden (2021-2023)). We can use a line plot.

```python
avg_bush = euro_to_dollar.loc[(euro_to_dollar['Time'] >= '2001-01-20') & (euro_to_dollar['Time'] <= '2009-01-20'),
                              'US_dollar'].mean()
avg_obama = euro_to_dollar.loc[(euro_to_dollar['Time'] > '2009-01-20') & (euro_to_dollar['Time'] <= '2017-01-20'),
                              'US_dollar'].mean()
avg_trump = euro_to_dollar.loc[(euro_to_dollar['Time'] > '2017-01-20') & (euro_to_dollar['Time'] <= '2021-01-20'),
                              'US_dollar'].mean()
avg_biden = euro_to_dollar.loc[(euro_to_dollar['Time'] > '2021-01-20'),
                              'US_dollar'].mean()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.text(dt.datetime(2003, 7, 1), 0.8, 'GEORGE W. BUSH', c='maroon',weight = 'bold')
ax.text(dt.datetime(2004, 5, 1), 0.77, '2001-2009', c='maroon',weight = 'bold',size=8)

ax.text(dt.datetime(2011, 8, 1), 0.8, 'BARACK OBAMA', c='darkorange',weight = 'bold')
ax.text(dt.datetime(2012, 6, 1), 0.77, '2009-2017', c='darkorange',weight = 'bold',size=8)

ax.text(dt.datetime(2017, 7, 1), 0.8, 'DONALD TRUMP', c='goldenrod',weight = 'bold')
ax.text(dt.datetime(2018, 5, 1), 0.77, '2017-2021', c='goldenrod',weight = 'bold',size=8)

ax.text(dt.datetime(2021, 7, 1), 0.8, 'JOE BIDEN', c='gold',weight = 'bold')
ax.text(dt.datetime(2021, 5, 1), 0.77, '2021-PRESENT', c='gold',weight = 'bold',size=8)

ax.text(dt.datetime(2001, 6, 1),1.55, 'Avg: '+ str(round(avg_bush,2)), c='maroon',
        bbox=dict(facecolor='none', edgecolor='maroon', boxstyle='round,pad=0.5'))
ax.text(dt.datetime(2009, 6, 1),1.55, 'Avg: '+ str(round(avg_obama,2)), c='darkorange',
        bbox=dict(facecolor='none', edgecolor='darkorange', boxstyle='round,pad=0.5'))
ax.text(dt.datetime(2017, 6, 1),1.55, 'Avg: '+ str(round(avg_trump,2)), c='goldenrod',
        bbox=dict(facecolor='none', edgecolor='goldenrod', boxstyle='round,pad=0.5'))
ax.text(dt.datetime(2021, 6, 1),1.55, 'Avg: '+ str(round(avg_biden,2)), c='gold',
        bbox=dict(facecolor='none', edgecolor='gold', boxstyle='round,pad=0.5'))


ax.plot(euro_to_dollar.loc[(euro_to_dollar['Time'] >= '2001-01-20') & (euro_to_dollar['Time'] <= '2009-01-20'),'Time'],
         euro_to_dollar.loc[(euro_to_dollar['Time'] >= '2001-01-20') & (euro_to_dollar['Time'] <= '2009-01-20'),'rolling_mean'],
         c='maroon',alpha=0.8,linewidth=1.9)
ax.plot(euro_to_dollar.loc[(euro_to_dollar['Time'] > '2009-01-20') & (euro_to_dollar['Time'] <= '2017-01-20'),'Time'],
         euro_to_dollar.loc[(euro_to_dollar['Time'] > '2009-01-20') & (euro_to_dollar['Time'] <= '2017-01-20'),'rolling_mean'],
         c='darkorange',alpha=0.8,linewidth=1.9)
ax.plot(euro_to_dollar.loc[(euro_to_dollar['Time'] > '2017-01-20') & (euro_to_dollar['Time'] <= '2021-01-20'),'Time'],
         euro_to_dollar.loc[(euro_to_dollar['Time'] > '2017-01-20') & (euro_to_dollar['Time'] <= '2021-01-20'),'rolling_mean'],
         c='goldenrod',alpha=0.8,linewidth=1.9)
ax.plot(euro_to_dollar.loc[(euro_to_dollar['Time'] > '2021-01-20'),'Time'],
         euro_to_dollar.loc[(euro_to_dollar['Time'] > '2021-01-20'),'rolling_mean'],
         c='gold',alpha=0.8,linewidth=1.9)

ax.axvline(dt.datetime(2009, 1, 20), color = 'black',alpha=0.3,linestyle ="dashed")
ax.axvline(dt.datetime(2017, 1, 20), color = 'black',alpha=0.3,linestyle ="dashed")
ax.axvline(dt.datetime(2021, 1, 20), color = 'black',alpha=0.3,linestyle ="dashed")

plt.ylim([0.7, 1.6])
plt.xlim([dt.datetime(2001, 1, 1), dt.datetime(2024, 1, 1)])

ax.set_xlabel('Year',size=12)
ax.set_ylabel('Exchange Rate',size=12)

ax.xaxis.label.set_color('dimgrey')        #setting up X-axis label color to yellow
ax.yaxis.label.set_color('dimgrey') 

ax.tick_params(axis='x', colors='dimgrey')    #setting up X-axis tick color to red
ax.tick_params(axis='y', colors='dimgrey')

ax.spines['left'].set_color('dimgrey')        # setting up Y-axis tick color to red
ax.spines['bottom'].set_color('dimgrey')

ax.spines[['right', 'top']].set_visible(False) # remove top and right border

ax.text(dt.datetime(2000, 4, 1), 1.65, 'EURO-USD rate during the terms of the last four US presidents',
        c='black',size = 13, weight= 'bold')

plt.show()
```
